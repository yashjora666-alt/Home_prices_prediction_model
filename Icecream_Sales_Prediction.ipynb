```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/icecream_sales.csv")
df
```

```python
X = df[['Temperature']]   # Independent variable
y = df['Sales']           # Dependent variable
```

```python
plt.scatter(X, y)
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.title("Temperature vs Ice Cream Sales")
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

```python
print("Slope (m):", model.coef_[0]) #Sales of Icecream per 1°C
print("Intercept (b):", model.intercept_) #Sales of Icecream if temprature is 0°C
```

```python
predicted_sales = model.predict([[50]])  # Example: 50°C
print(predicted_sales)
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.title("Regression Line")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x = df[["Temperature"]]
y = df[["Sales"]]
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2) #80% would be for training and 20% for testing
```

```python
model.fit(train_x,train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
print("coefficient (slope):", model.coef_[0])
print("intercept:", model.intercept_)
print("MSE:", mean_squared_error(test_y, y_pred)) # Lower = Better
print("R2 score:", r2_score(test_y, y_pred))
```
